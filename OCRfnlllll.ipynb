```python
from PIL import Image
import torch
from transformers import NougatProcessor, VisionEncoderDecoderModel

# Specify the directory where the model and processor were saved
save_directory = "/content/Fnlocr"

# Load the processor and model from the saved directory
processor = NougatProcessor.from_pretrained(save_directory)
model = VisionEncoderDecoderModel.from_pretrained(
    save_directory,
    torch_dtype=torch.bfloat16,
)

# Get the max context length of the model & dtype of the weights
context_length = model.decoder.config.max_position_embeddings
torch_dtype = model.dtype

# Move the model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def predict(img_path):
    # Prepare PDF image for the model
    image = Image.open(img_path)
    pixel_values = (
        processor(image, return_tensors="pt").pixel_values.to(torch_dtype).to(device)
    )

    # Generate transcription
    outputs = model.generate(
        pixel_values.to(device),
        repetition_penalty=1.5,
        min_length=1,
        max_new_tokens=context_length,
        bad_words_ids=[[processor.tokenizer.unk_token_id]],
    )

    page_sequence = processor.batch_decode(outputs, skip_special_tokens=True)[0]
    return page_sequence

# Run model and store result
model_output = predict("/content/982abef974af643202ff567b811deec4.jpg")

# Debug print to see the output
print("===== Model Output =====")
print(model_output)
```

```python
import json
import re
from difflib import SequenceMatcher

# Load Quran JSON
with open('/content/quran.json', 'r', encoding='utf-8') as f:
    quran = json.load(f)

# Flatten JSON: extract all ayahs and metadata
all_ayahs = []
ayah_refs = []  # (surah, verse, text)

for surah_ayahs in quran.values():
    for verse in surah_ayahs:
        all_ayahs.append(verse['text'])
        ayah_refs.append((verse['chapter'], verse['verse'], verse['text']))

# Raw model output (replace this with real output)
raw_model_output = model_output

# Clean model output
cleaned_model = re.sub(r"[\[\]#()«»<>]", "", raw_model_output)
cleaned_model = re.sub(r"http\S+", "", cleaned_model)
cleaned_lines = [line.strip() for line in cleaned_model.split("\n") if line.strip()]

# Match helper with reference return
def find_closest_match(line, ayah_refs, threshold=0.5):
    best_match = None
    best_ratio = 0
    for chapter, verse, text in ayah_refs:
        ratio = SequenceMatcher(None, line, text).ratio()
        if ratio > best_ratio:
            best_ratio = ratio
            best_match = (chapter, verse, text)
    return best_match if best_ratio > threshold else ("NO MATCH", "", line)

# Match each model line
matched_output = []
for line in cleaned_lines:
    match = find_closest_match(line, ayah_refs)
    matched_output.append(match)

# Print output
print("==== Raw Model Output ====")
print(raw_model_output.strip())

print("\n==== Final Matched Output (Surah:Ayah) ====")
for chapter, verse, ayah_text in matched_output:
    if chapter == "NO MATCH":
        print(f"[NO MATCH] → {ayah_text}")
    else:
        print(f"{chapter}:{verse} → {ayah_text}")
```
